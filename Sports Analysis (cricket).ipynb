#### Q. 13)  Create table deliveries_v02 with all the columns of the table â€˜deliveriesâ€™ and an additional column ball_result containing values boundary, dot or other depending on the total_run (boundary for >= 4, dot for 0 and other for any other number)(Hint 1 : CASE WHEN statement is used to get condition based results) (Hint 2: To convert the output data of select statement into a table, you can use a subquery. Create table table_name as [entire select statement].

```python
query = """ create table deliveries_v02 as select *, 
            CASE WHEN total_runs = 4 then 'boundary'
            WHEN total_runs = 6 then 'six'
            WHEN total_runs = 0 THEN 'duck' 
            ELSE 'other' 
            END as ball_result 
            FROM deliveries """
mydb.commit()
cursor = mydb.cursor()
cursor.execute(query)
cursor.fetchall()
print(cursor.rowcount,"records are inserted")
#print("Query exicuted")
```

```python
pd.read_sql('select * from deliveries_v02 limit 10',mydb)
```

#### Q.15)  Write a query to fetch the total number of boundaries or sixes scored by each team from the deliveries_v02 table and order it in descending order of the number of boundaries scored.

```python
# IN 
mydb.commit()
cursor = mydb.cursor()
query = """ 
select batting_team, count(*) as boundaries_scored
from deliveries_v02 where ball_result IN ('boundary','six')
group by batting_team order by count(*) desc
"""
pd.read_sql(query,mydb)
```

#### Q. 16) Write a query to fetch the total number of dot balls bowled by each team and order it in descending order of the total number of dot balls bowled.

```python
# Order by 
query = """
select bowling_team, count(*) as 'dot_balls_played' from deliveries_v02 
where ball_result = 'duck' group by bowling_team 
order by count(*) desc
"""
pd.read_sql(query,mydb)
```

#### Q. 17) Write a query to fetch the total number of dismissals by dismissal kinds where dismissal kind is  NA.

#### Q. 20) Write a query to create a table named deliveries_v03 with all the columns of deliveries_v02 table and two additional column (named venue and match_date) of venue and date from table matches.

```python
query = """
CREATE TABLE deliveries_v03 AS SELECT a.*, b.venue, b.match_date from 
deliveries_v02 as a 
left join (select venue,date as match_date, id from matches group by 
id) as b
on a.id = b.id;

"""
mydb.commit()
cursor = mydb.cursor()
cursor.execute(query)
cursor.fetchall()
print(cursor.rowcount,"records are inserted")
#print("Query exicuted")
```

```python
pd.read_sql("SELECT * FROM deliveries_v03 limit 10 ",mydb)
```

#### Q.26) Find Virat Kohli's Performance against all teams.

```python
a = pd.read_sql(""" SELECT batsman,bowling_team, SUM(batsman_runs) AS 'Total_runs', 
                COUNT(batsman_runs)as 'balls_played',ROUND((SUM(batsman_runs)/COUNT(batsman_runs))*100,2) AS 'Strike_rate'
                FROM deliveries
                WHERE Batsman = "V Kohli"
                GROUP BY bowling_team 
                ORDER BY Total_runs DESC  """,mydb)
a
```

#### Q.27) Top 10 batsman with centuries in IPL.

```python
a = pd.read_sql(""" SELECT Batsman,ID, SUM(batsman_runs) AS 'Total_runs' 
                FROM deliveries
                GROUP BY Batsman,ID
                HAVING Total_runs >= 100
                ORDER BY Total_runs DESC limit 10""",mydb)
a
```

#### Q.28) find the top 5 batsman with highest strike rate who have played a min 1000 balls.

```python
a = pd.read_sql(""" SELECT Batsman, SUM(batsman_runs) AS 'Total_runs', COUNT(batsman_runs),
                ROUND((SUM(batsman_runs)/COUNT(batsman_runs))*100,2) AS 'Strike_rate'
                FROM deliveries
                GROUP BY Batsman
                HAVING COUNT(batsman_runs) > 1000
                ORDER BY Strike_rate DESC LIMIT 10 """,mydb)
a
```
