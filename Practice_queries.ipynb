```python
# Finding the average of balance amount 
pd.read_sql("select AVG(balance) as Average_balance from bank_details",mydb)
```

```python
# finding out who is having minimum balance 
#First method
pd.read_sql('select job,min(balance) as low_balance from bank_details groupby having min(balance)',mydb)
```

```python
# second method 
pd.read_sql("select job,min(balance) from bank_details",mydb)
```

```python
# finding out who is having maximum balance 
# first method 
pd.read_sql('select job,max(balance) from bank_details groupby having max(balance)',mydb)
```

```python
#second method 
pd.read_sql("select job,max(balance) from bank_details",mydb)
```
